1 - Génération de données
---

Dans cette première partie nous allons générer des données selon 3 Gaussiennes sphériques. Vous utiliserez pour cela la fonction `np.random.normal`. Les caractéristiques des 3 gausiennes sont les suivantes:
- 500 points selon une distribution gaussienne centrée sur le point [0,2], d'écart type 0.1,
- 100 points selon une distribution gaussienne centrée sur le point [-1,0], d'écart type 0.05
- 50  points selon une distribution gaussienne centrée sur le point [1,0], d'écart type 0.05

```python
## L'écart type (ou déviation standard)mesure la dispersion des valeurs par rapport à la moyenne.
## l'écart type (σ) est défini comme la racine carrée de la variance (σ^2),
## où la variance est la moyenne des carrés des écarts entre chaque valeur et la moyenne.
def gen_3_gausiennes():
    '''
    :return: retourne trois matrices correspondant au trois gausiennes dans l'ordre de la question
    '''
    g1 = np.random.normal(loc = [0,2], scale = 0.1, size=(500, 2))
    g2 = np.random.normal(loc = [-1,0], scale = 0.05, size=(100, 2))
    g3 = np.random.normal(loc = [1,0], scale = 0.05, size=(50, 2))
    
    return g1,g2,g3
```

```python
# Ce bloc permet de valider votre code. Vous ne devez pas le modifier.  
try:
    g1,g2,g3 = gen_3_gausiennes()
    np.testing.assert_equal(g1.shape,(500,2),err_msg="\033[93m {}\033[00m" .format('Test 1 : le nombre de point généré par la première gaussienne est incorrect.'))
    np.testing.assert_equal(g2.shape,(100,2),err_msg="\033[93m {}\033[00m" .format('Test 2 : le nombre de point généré par la deuxième gaussienne est incorrect.'))
    np.testing.assert_equal(g3.shape,(50,2),err_msg="\033[93m {}\033[00m" .format('Test 3 : le nombre de point généré par la troisième gaussienne est incorrect.'))
        
except Exception as e: 
    print("\033[91m {}\033[00m" .format('KO - Au moins un test n\'est pas validé')) 
    print('Information sur le test non valide:')
    print(e)
    raise e
else:
    print("\033[92m {}\033[00m" .format('Ok - Tous les tests sont validés.')) 
    

del g1,g2,g3
```

Écrivez une fonction qui génére les points selon les trois gausiennes et les concaténe dans une seule matrice. Pour rappel les fonctions de concaténation de `numpy` sont listé à l'adresse : https://numpy.org/doc/stable/reference/routines.array-manipulation.html#joining-arrays

```python
def gen_pts():
    # Generate the three Gaussian distributions
    g1, g2, g3 = gen_3_gausiennes()
    X = np.concatenate((g1, g2, g3), axis=0)
    return X

X = gen_pts()
##np.set_printoptions(threshold=np.inf)
##print(X)
```

2 - Clustering par K-Moyennes
---

Nous allons dans cette partie utiliser l'implémentation de `scikit-learn` de l'algoritheme des K-Moyennes et en étudier les paramètres. 
Commencez par lire la documentation sur le sujet : https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

L'algorithme des K-Moyennes peut être initialisé de plusieurs façon. Faites un apprentissage en partant des trois points suivants : 
- -0.4  ;  2.75
- 0 ;  2.75
- 0.3  ; 2.5

Vous mettrez le paramètre `n_init` à 1.

```python
from sklearn.cluster import KMeans
def get_K_Means_with_spetial_init(X):
    '''
    :param X: données uitilisées pour apprendre le Kmeans
    :return: retourne l'objet appris à l'aide d'un Kmeans sur les données X passé en argument de la fonction. L'algorithme est initialisé avec 3 centres fixés.
    '''
    kmeans = KMeans(n_clusters=3,init='k-means++', n_init=1)
    
    kmeans.fit(X)
    
    return kmeans
```

```python
# Get the K-Means model with special initialization
kmeans_model = get_K_Means_with_spetial_init(X)

# Access the cluster centers
cluster_centers = kmeans_model.cluster_centers_
print("Cluster Centers:", cluster_centers)

# Access the cluster labels assigned to each data point
labels = kmeans_model.labels_
print("Cluster Labels:", labels)
```

```python
# Ce bloc permet de valider votre code. Vous ne devez pas le modifier.  
try:
    _kmeans_ = get_K_Means_with_spetial_init(X)
    
    np.testing.assert_equal(type(_kmeans_),sklearn.cluster._kmeans.KMeans,err_msg="\033[93m {}\033[00m" .format('Test 1 : l\'objet retourné par la fonction get_K_Means_with_spetial_init n\'est pas correct.'))
    try:
        assert 'n_iter_' in _kmeans_.__dict__
    except Exception as e: 
        raise AssertionError("Test 2 : L'algorithme Kmeans n'a pas été appris.")
except Exception as e: 
    print("\033[91m {}\033[00m" .format('KO - Au moins un test n\'est pas validé')) 
    print('Information sur le test non valide:')
    print(e)
    raise e
else:
    print("\033[92m {}\033[00m" .format('Ok - Tous les tests sont validés.')) 
    
```

Affichez sur une même figure:
- les points dans `X`
- les 3 points de départs
- la position finales des 3 clusters.

Vous utiliserez des couleurs différentes pour différencier ces points.

```python
def plot_kmeans_with_spetial_init(X):
    
    kmeans_model = get_K_Means_with_spetial_init(X)

    cluster_centers_initial = np.array([[-0.4, 2.75], [0, 2.75], [0.3, 2.5]])
    
    cluster_centers_final = kmeans_model.cluster_centers_
    
    labels = kmeans_model.labels_

    # Plot the data points with cluster colors
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Data Points')

    # Plot the initial points
    plt.scatter(cluster_centers_initial[:, 0], cluster_centers_initial[:, 1], c='blue', marker='o', s=20, label='Initial Points')

    # Plot the final cluster centers
    plt.scatter(cluster_centers_final[:, 0], cluster_centers_final[:, 1], c='red', marker='X', s=20, label='Final Cluster Centers')

    # Add labels and legend
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('K-Means Clustering with Special Initialization')
    plt.legend()

    # Show the plot
    plt.show()

# 10 clustering avec des points différents. Que pouvez-vous remarquer ? Pourquoi ? 
for i in range(10):
    X = gen_pts()
    plot_kmeans_with_spetial_init(X)
```

Exécutez le code suivant pour visualiser les cellules de Voronoï de chaque cluster.

```python
kmeans = get_K_Means_with_spetial_init(X)

def affichage_cellules_Voronoi(clf,X,labels):
    '''
    Fonction affichant les cellules de Voronoi d'un algorithme de clustering préalablement appris
    :param clf: classifieur non supervisé appris au préalable
    '''
    pas = 500 # pas de quantification, plus la valeur est grande plus les frontières sont précises mais le temps de calcul augmente

    x_min, x_max = X[:, 0].min() -1, X[:, 0].max() +1
    y_min, y_max = X[:, 1].min() -1 , X[:, 1].max() +1
    h = abs(x_max-x_min)/pas 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = clf.predict(np.stack((xx.ravel(), yy.ravel()),1))
    Z = Z.reshape(xx.shape)

    plt.scatter(X[:,0],X[:,1],c=labels,marker='.')
    plt.imshow(Z, interpolation='nearest',
               extent=(xx.min(), xx.max(), yy.min(), yy.max()),
               cmap=plt.cm.Paired,
               aspect='auto', origin='lower')
    plt.show()
    
affichage_cellules_Voronoi(kmeans,X,labels)
```

3 - Modèle de mélange de Gausiennes

Nous allons dans cette partie réalisé un clustering à l'aide d'un mélange de Gausinnes (GMM: Gaussian Mixture Model). Pour cela veuillez étudiez la documentation à la page: https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html

De manière analogue à la partie précédente. Apprenez le mélange de Gaussiennes et affichez les cellules de Voronoï de chaque cluster.

```python
from sklearn.mixture import GaussianMixture
def get_GMM(X):
    '''
    :param X: données d'apprentissage
    :return : un GMM appris avec X
    '''
    gmm = GaussianMixture(n_components=3, n_init=1, init_params='kmeans')
    
    gmm.fit(X)
    return gmm

gmm = get_GMM(X)
affichage_cellules_Voronoi(gmm,X,labels)
```

Les données que nous utilisons jusqu'à présent sont définis selon des clusters sphériques. Nous allons maintenant utiliser des données réparties selon des ellipses. Pour cela utilisez la fonction `np.random.multivariate_normal` ( https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html ).

Les caractéristiques des 3 lois normales sont les suivantes:
- 500 points selon une distribution normal centrée sur le point [4,6], de covariance $\left[\begin{array}{cc}
1&0.7\\
0.7&1
\end{array}
\right]
$,
- 100 points selon une distribution normal centrée sur le point [0,0], de covariance $\left[\begin{array}{cc}
0.7&-0.5\\
-0.5&0.7
\end{array}
\right]
$,
- 50  points selon une distribution normal centrée sur le point [7,2], de covariance $\left[\begin{array}{cc}
0.3&0\\
0&0.3
\end{array}
\right]
$.

```python
def gen_pts_3cluster_ellipse():
    '''
    :return : des points générer selon trois distributions normals elliptiques pré-définis.
    '''
    x1 = np.random.multivariate_normal([4,6], [[1,0.7],[0.7,1]], size = 500)
    x2 = np.random.multivariate_normal([0,0], [[0.7,-0.5],[-0.5,0.7]] , size = 100)
    x3 = np.random.multivariate_normal([7,2], [[0.3,0],[0,0.3]], size = 50)
    
    X = np.concatenate((x1, x2, x3), axis=0)
    
    return X

X_ellipse = gen_pts_3cluster_ellipse()
```

```python
from sklearn.cluster import SpectralClustering

def get_spectral_clustering(X):
    '''
    :param X: données d'apprentissage
    :return : un clustering spectral sur les données X
    '''
    
    spectre = SpectralClustering(n_clusters=3, eigen_solver="arpack", affinity='nearest_neighbors').fit_predict(X)
    
    return spectre
```

L'objectif de cette question est de recoder une fonction analogue à celle que vous avez utiliser avec `scikit-learn`. Vous n'êtes donc pas autorisé à utiliser la librairie `scikit-learn` dans cette partie. Votre solution ne doit utiliser que les libraires `numpy` et `scipy`.

Écrivez une fonction *initCentre* qui prend le nombre de centre en paramètre, le nombre de dimension des centres (2 dans les exemples précédent) et retourne des centres tirés selon une loi uniforme entre 0 et 1. Cela correspondra à l’initialisation de l’algorithme.

```python
import numpy as np

def initCentre(n,d):
    '''
    :param n: nombre de centre
    :param d: dimension de chaque centre
    :return : Retourne n centre de dimension d initialisé aléatoirement selon une loi uniforme
    '''
    centers = np.random.uniform(low=0.0, high=1.0, size=(n,d))
    
    return centers
```

Faites une fonction `association` qui prend en argument des points et des centres et qui retourne l’association entre chaque point et le numéro du centre le plus proche. Pour réaliser cette fonction, vous pouez utiliser la fonction `cdist` de `scipy` (https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cdist.html) . 

```python
import scipy.spatial.distance as ds
def association(p,c):
    '''
    :param p: des points
    :param c: des centres
    :return: l'association de chaque point au centre le plus proch.
    '''
    distances = ds.cdist(p, c, metric='euclidean')
    associations = np.argmin(distances, axis=1)
    return associations
```

Créez une fonction *kmeans* implémentant l’algorithme de K-moyennes. Cette fonction prendra comme argument les points d’apprentissage, le nombre de cluster recherché et le nombre d’itération de l’algorithme. Elle retournera les centres trouvés. Vous utiliserez une boucle for pour chaque itération ainsi que les fonctions : *initCentre*, *association* et *updateCentres*.

```python
def kmeans(x, nbCenter, nIteration):
    """
    Implémente l'algorithme de K-means.

    Parameters:
    - x (numpy.ndarray): Un tableau 2D représentant les coordonnées des points.
    - nbCenter (int): Le nombre de clusters recherché.
    - nIteration (int): Le nombre d'itérations de l'algorithme.

    Returns:
    - centres (numpy.ndarray): Un tableau 2D contenant les centres trouvés.
    """
    centres = initCentre(nbCenter, x.shape[1])

    for _ in range(nIteration):
        associations = association(x, centres)
        centres = updateCentres(x, associations, nbCenter)

    return centres
```

```python
import numpy as np

def calculate_distance_matrix_numpy(data):
    """
    Calculate the distance matrix between all pairs of points in the dataset.
    """
    num_points = data.shape[0]
    distance_matrix = np.zeros((num_points, num_points))
    for i in range(num_points):
        for j in range(i, num_points):
            distance_matrix[i, j] = np.linalg.norm(data[i] - data[j])
            distance_matrix[j, i] = distance_matrix[i, j]  # Distance matrix is symmetric
    return distance_matrix
```

```python
def calculate_distance_matrix(data):
    """
    Calculate the distance matrix between all pairs of points in the dataset.
    """
    distance_matrix = cdist(data, data)
    return distance_matrix
```

```python
def find_core_points(distance_matrix, eps, min_samples):
    """
    Find core points in the dataset based on epsilon and min_samples.
    """
    num_points = distance_matrix.shape[0]
    core_points = []
    for i in range(num_points):
        if np.sum(distance_matrix[i] <= eps) >= min_samples:
            core_points.append(i)
    return core_points
```

```python
def expand_cluster(point_index, neighbors, cluster_labels, cluster_id, eps, min_samples, distance_matrix):
    """
    Expand the cluster starting from a core point.
    """
    cluster_labels[point_index] = cluster_id
    i = 0
    while i < len(neighbors):
        neighbor_index = neighbors[i]
        if cluster_labels[neighbor_index] == -1:
            cluster_labels[neighbor_index] = cluster_id
        elif cluster_labels[neighbor_index] == 0:
            cluster_labels[neighbor_index] = cluster_id
            neighbor_neighbors = np.where(distance_matrix[neighbor_index] <= eps)[0]
            if len(neighbor_neighbors) >= min_samples:
                neighbors = np.concatenate((neighbors, neighbor_neighbors))
        i += 1
```

```python
def dbscan(data, eps, min_samples):
    """
    Perform DBSCAN clustering on the given dataset.
    """
    distance_matrix = calculate_distance_matrix(data)
    cluster_labels = np.zeros(len(data), dtype=int)  # 0 indicates unclassified
    cluster_id = 0
    for i in range(len(data)):
        if cluster_labels[i] == 0:
            neighbors = np.where(distance_matrix[i] <= eps)[0]
            if len(neighbors) < min_samples:
                cluster_labels[i] = -1  # Noise point
            else:
                cluster_id += 1
                expand_cluster(i, neighbors, cluster_labels, cluster_id, eps, min_samples, distance_matrix)
    return cluster_labels
```

```python
import numpy as np
from scipy.sparse import csgraph
from scipy.linalg import eigh

def construct_similarity_graph(data, sigma=1.0, k=None):
    """
    Construct a similarity graph based on the input data.
    """
    # Calculate pairwise distances
    pairwise_distances = np.linalg.norm(data[:, None] - data, axis=-1)
    
    # Compute Gaussian similarity
    if sigma is not None:
        similarity_matrix = np.exp(-pairwise_distances ** 2 / (2 * (sigma ** 2)))
    # Alternatively, compute k-nearest neighbors similarity
    elif k is not None:
        similarity_matrix = np.zeros_like(pairwise_distances)
        for i in range(len(data)):
            indices = np.argsort(pairwise_distances[i])[:k+1]  # Include the point itself
            similarity_matrix[i, indices] = 1
    else:
        raise ValueError("Either sigma or k must be provided.")
    
    # Make the similarity matrix symmetric
    similarity_matrix = 0.5 * (similarity_matrix + similarity_matrix.T)
    
    return similarity_matrix

def compute_laplacian(similarity_matrix, normalization='symmetric'):
    """
    Compute the graph Laplacian matrix.
    """
    laplacian_matrix = csgraph.laplacian(similarity_matrix, normed=(normalization == 'symmetric'))
    return laplacian_matrix

def spectral_embedding(laplacian_matrix, n_components):
    """
    Perform spectral embedding using eigenvalue decomposition.
    """
    eigenvalues, eigenvectors = eigh(laplacian_matrix)
    embedding = eigenvectors[:, :n_components]
    return embedding

def k_means_clustering(embedding, n_clusters, max_iter=300):
    """
    Perform k-means clustering in the embedded space.
    """
    # Initialize centroids randomly
    centroids = embedding[np.random.choice(embedding.shape[0], n_clusters, replace=False)]
    for _ in range(max_iter):
        # Assign each point to the nearest centroid
        labels = np.argmin(np.linalg.norm(embedding[:, None] - centroids, axis=-1), axis=-1)
        # Update centroids
        new_centroids = np.array([embedding[labels == k].mean(axis=0) for k in range(n_clusters)])
        # Check for convergence
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
    return labels

def spectral_clustering(data, n_clusters, sigma=1.0, k=None, normalization='symmetric', n_components=None):
    """
    Perform Spectral Clustering on the input data.
    """
    similarity_matrix = construct_similarity_graph(data, sigma=sigma, k=k)
    laplacian_matrix = compute_laplacian(similarity_matrix, normalization=normalization)
    if n_components is None:
        n_components = n_clusters
    embedding = spectral_embedding(laplacian_matrix, n_components)
    labels = k_means_clustering(embedding, n_clusters)
    return labels
```

```python
from scipy.stats import multivariate_normal

class GaussianMixtureModel:
    def __init__(self, n_components=1, max_iter=100, tol=1e-6, random_state=None):
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol
        self.random_state = random_state
        self.weights = None
        self.means = None
        self.covariances = None
        self.log_likelihoods = []

    def fit(self, X):
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape
        
        # Initialize parameters
        self.weights = np.ones(self.n_components) / self.n_components
        self.means = X[np.random.choice(n_samples, self.n_components, replace=False)]
        self.covariances = [np.eye(n_features)] * self.n_components

        for _ in range(self.max_iter):
            # Expectation step
            gamma = self._expectation(X)
            
            # Maximization step
            self._maximization(X, gamma)
            
            # Calculate log-likelihood
            log_likelihood = self._log_likelihood(X)
            self.log_likelihoods.append(log_likelihood)
            
            # Check for convergence
            if len(self.log_likelihoods) >= 2 and np.abs(log_likelihood - self.log_likelihoods[-2]) < self.tol:
                break

    def _expectation(self, X):
        n_samples = X.shape[0]
        gamma = np.zeros((n_samples, self.n_components))

        for k in range(self.n_components):
            gamma[:, k] = self.weights[k] * multivariate_normal.pdf(X, mean=self.means[k], cov=self.covariances[k])

        gamma /= np.sum(gamma, axis=1, keepdims=True)
        return gamma

    def _maximization(self, X, gamma):
        n_samples, _ = X.shape
        total_weight = np.sum(gamma, axis=0)

        # Update weights
        self.weights = total_weight / n_samples

        # Update means
        self.means = np.dot(gamma.T, X) / total_weight[:, np.newaxis]

        # Update covariances
        for k in range(self.n_components):
            diff = X - self.means[k]
            self.covariances[k] = np.dot(gamma[:, k] * diff.T, diff) / total_weight[k]

    def _log_likelihood(self, X):
        log_likelihood = 0
        for k in range(self.n_components):
            log_likelihood += self.weights[k] * multivariate_normal.pdf(X, mean=self.means[k], cov=self.covariances[k])
        return np.sum(np.log(log_likelihood))

    def predict(self, X):
        gamma = self._expectation(X)
        return np.argmax(gamma, axis=1)
```
